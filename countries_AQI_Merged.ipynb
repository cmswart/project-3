```python
#dependencies and set up
import pandas as pd
from pathlib import Path
```

```python
#load csv into dataframe
countries_df = pd.read_csv("countries.csv")

#display data
countries_df.head()
```

```python
#Removing space after each country name
clean_df = countries_df.replace({'Country': {'United States ': 'United States', 'Canada ': 'Canada', 'France ': 'France', 'Germany ': 'Germany', 'Italy ': 'Italy', 'Japan ': 'Japan', 'United Kingdom ': 'United Kingdom'}})
#new_df = countries_df.replace({"Canada ", "Canada"})
#new_df = countries_df.replace({"France ", "France"})
#new_df = countries_df.replace({"Germany ", "Germany"})
#new_df = countries_df.replace({"Italy ", "Italy"})
#new_df = countries_df.replace({"Japan ", "Japan"})
#new_df = countries_df.replace({"United Kingdom ", "United Kingdom"})


clean_df
```

```python
countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']

filtered_one_df = clean_df['Country'].isin(countries)

final_df = clean_df[filtered_one_df]

final_df 
#filtered_df =  countries_df.loc[(countries_df['Country'] == 'Canada ') | (countries_df['Country'] == 'France ') | 
                                 #(countries_df['Country'] == 'Germany ') | (countries_df['Country'] == 'Italy ') | 
                                 #(countries_df['Country'] == 'Japan ') | (countries_df['Country'] == 'United Kingdom ') | 
                                 #(countries_df['Country'] == 'United States '), :]

#filtered_df
```

```python
# Load the CSV file into a Pandas DataFrame
aqi_data_df = pd.read_csv("AQI_data.csv")

# Display sample data
aqi_data_df
```

```python
# Filter for countries(Canada, France, Germany, Italy, Japan, United Kingdom, United States)
countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom of Great Britain and Northern Ireland', 'United States of America']

G7_AQI_Countries =  aqi_data_df[aqi_data_df['Country'].isin(countries)]

G7_AQI_Countries = G7_AQI_Countries.replace({'Country': {'United States of America': 'United States', 'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom'}})

G7_AQI_Countries
```

```python
# Merge two DataFrames
merge_df = pd.merge(final_df, G7_AQI_Countries, on=["Country"])
merge_df
```

```python
# Save Merged DataFrames to CSV File
merge_df.to_csv("Merge_AQI.csv",index=False)
```

```python
#Average CO AQI Value for Each Country
merge_df.groupby('Country') ['CO AQI Value'].mean()
```

```python
#Average Ozone AQI Value for Each Country
merge_df.groupby('Country') ['Ozone AQI Value'].mean()
```

```python
#Average NO2 AQI Value for Each Country
merge_df.groupby('Country') ['NO2 AQI Value'].mean()
```

```python
#Average PM2.5 AQI Value
merge_df.groupby('Country') ['PM2.5 AQI Value'].mean()
```

```python
avg_df = merge_df.groupby('Country') ['Ozone AQI Value','NO2 AQI Value','PM2.5 AQI Value', 'CO AQI Value'].mean()

json_file = avg_df.to_json(orient='table')
with open('G7_AQI_Countries.json', 'w') as f:
    f.write(json_file)
avg_df
```
